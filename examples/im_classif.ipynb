## Image Classification Example

This notebook imports the [`im_classif.py`](https://github.com/forman/dectree/blob/master/examples/im_classif.py) module which has been generated from the decision tree in [`im_classif.yml`](https://github.com/forman/dectree/blob/master/examples/im_classif.yml) using the following command:

```bash
$ dectree examples/im_classif.yml --vectorize
```

````python
import im_classif
````

````python
from PIL import Image
import numpy as np
````

### Import a test image

````python
im = Image.open('test_im.jpg')
````

````python
im
````

### Extract the R,G,B channels as Numpy arrays of type `float64`

````python
red = np.array(im.getdata(band=0), dtype=np.float64)
green = np.array(im.getdata(band=1), dtype=np.float64)
blue = np.array(im.getdata(band=2), dtype=np.float64)
````

### Classification input

````python
input = im_classif.Input()
input.red = red
input.green = green
input.blue = blue
````

### Classification output

````python
output = im_classif.Output()
````

### Classify

````python
im_classif.apply_rules(input, output)
````

### Display outputs

````python
def truth_to_image(truth):
    im_data = np.array(255 * truth, dtype=np.uint8).reshape(im.height, im.width)
    return Image.fromarray(im_data)
````

````python
truth_to_image(output.grey)
````

````python
truth_to_image(output.yellow)
````

````python
truth_to_image(output.dark_red)
````

````python
truth_to_image(output.dark)
````

````python
truth_to_image(output.not_dark)
````
